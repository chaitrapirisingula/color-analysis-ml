```python
from ColorBlindnessAnalysis import *
from ColorMasks import *
import cv2
```

```python
cb_masks = [protanopia_simulation_matrix, deuteranopia_simulation_matrix, tritanopia_simulation_matrix]
cb_names = ["Protanopia", "Deuteranopia", "Tritanopia"]
```

```python
# Load the original image
img = cv2.imread('images/legos.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Get original colors
original_colors = get_colors(img, 20, "Original")
```

```python
# Simulate Protanopia on original image
protanopia_image = simulate_colorblindness(protanopia_simulation_matrix, img)
protanopia_colors = get_colors(protanopia_image, 20, cb_names[0])

# Recolor to improve contrast for protanopia
recolored_protanopia = improve_contrast_red_green(img)

# Display new image
protanopia_image = simulate_colorblindness(protanopia_simulation_matrix, recolored_protanopia)
protanopia_colors = get_colors(protanopia_image, 20, "Recolored " + cb_names[0])
```

```python
# Simulate Protanopia on original image
tritanopia_image = simulate_colorblindness(tritanopia_simulation_matrix, img)
tritanopia_colors = get_colors(tritanopia_image, 20, cb_names[2])

# Recolor to improve contrast for tritanopia
recolored_tritanopia = improve_contrast_blue_red(img)

# Display new image
tritanopia_image = simulate_colorblindness(tritanopia_simulation_matrix, recolored_tritanopia)
tritanopia_colors = get_colors(tritanopia_image, 20, "Recolored " + cb_names[2])
```
